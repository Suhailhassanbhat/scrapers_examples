```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = 'https://klec.ky.gov/Reports/Pages/Legislators-and-Candidates.aspx'
response = requests.get(url, verify=False).content
```

```python
doc=BeautifulSoup(response, 'html.parser')
```

```python
senate_members = []
for z in doc.find_all(class_="row-fluid")[0].find_all('li'):
    for x in z.find_all('a'):
#         print(x['href'], x.text)
        row = {}
        row['senate_member']=x.text
        row['disclosure_link'] = "https://klec.ky.gov"+x['href']
        
        senate_members.append(row)

df=pd.DataFrame(senate_members)
df
df.to_excel('senate_members_ky.xlsx', index=False)
```

```python
congress_members = []
for z in doc.find_all(class_="row-fluid")[1].find_all('li'):
    for x in z.find_all('a'):
#         print(x['href'], x.text)
        row = {}
        row['congress_member']=x.text
        row['disclosure_link'] = "https://klec.ky.gov"+x['href']
        
        congress_members.append(row)

df=pd.DataFrame(congress_members)
df
df.to_excel('congress_members_ky.xlsx', index=False)
```
