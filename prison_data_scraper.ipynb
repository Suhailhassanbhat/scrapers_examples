```python
import os
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup
import pandas as pd
import ssl
import re
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# connect to website and get list of all pdfs
url="https://corrections.ky.gov/About/researchandstats/Pages/dailycount.aspx"
response = requests.get(url, verify=False)
soup= BeautifulSoup(response.text, "html.parser")     
```

```python
# #If there is no such folder, the script will create one automatically
# folder_location = r'C:\Users\suhailbhat\Desktop\prison_data'
# if not os.path.exists(folder_location):os.mkdir(folder_location)
```

```python
links = soup.find_all('a', href=re.compile(r'(.pdf)'))
```

```python
# clean the pdf link names
url_list = []
for el in links:
    if(el['href'].startswith('http')):
        url_list.append(el['href'])
    else:
        url_list.append("https://corrections.ky.gov" + el['href'])

print(url_list)
```

```python
# download the pdfs to a specified location
for url in url_list:
    filename = os.path.join(url.replace("https://corrections.ky.gov/", ""))
    print(filename) 
    print(url)
```

```python
# connect to website and get list of all pdfs
url="https://corrections.ky.gov/About/researchandstats/Pages/WeeklyJail.aspx"
response = requests.get(url, verify=False)
soup= BeautifulSoup(response.text, "html.parser")    
```

```python
links = soup.find_all('a', href=re.compile(r'(.pdf)'))
```

```python
# clean the pdf link names
url_list = []
for el in links:
    row={}
    row['filename'] =  el['href']
    row['url']= "https://corrections.ky.gov" + el['href']
    
    url_list.append(row)
    
#     if(el['href'].startswith('https')):
#         url_list.append(el['href']).strip()
#     else:
#         url_list.append("https://corrections.ky.gov" + el['href'])

url_list
```

```python
df = pd.DataFrame(url_list)
df = df.dropna( how='all')
```

```python
df['cleaned_name']=df['filename'].str.replace("/about/researchandstats/documents/weekly jail/", "")
df['cleaned_name'] = df['cleaned_name'].str.replace("/About/researchandstats/Documents/Weekly%20Jail/", "")
```

```python
df=df[["cleaned_name", "url"]]
```

```python
df=df.head(120)
```

```python
for urls in df['url']:
    link = urls
    book_name = "/Users/suhailbhat/Desktop/jail_data/" + link.split('/')[-1]
    with open(book_name, 'wb') as book:
        a = requests.get(link, stream=True, verify=False)

        for block in a.iter_content(512):
            if not block:
                break

            book.write(block)
```
